## 3.1.3. 3차 함수 그리기


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def fnTmp(x):
    return (x - 2) * x * (x + 2)

print(fnTmp(1))
```

## 3.1.6. 그래프 장식

범례, 축범위, 제목, 라벨

```python
def fnQuadratic(x, w):
    return (x - w) * x * (x + 2) 

# x 정의
x = np.linspace(-3, 3, 100)

# 그래프
plt.plot(x, fnQuadratic(x, 2),
         color = "black",
         label = "$w = 2$")
plt.plot(x, fnQuadratic(x, 1),
         color = "cornflowerblue",
         label = "$w = 1$")
plt.legend(loc = "upper left") # 범례 표시
plt.ylim(-15, 15)              # y축 범위
plt.title("$f_2(x)$")          # 제목
plt.xlabel("$x$")              # x축 라벨
plt.ylabel("$y$")              # y축 라벨
plt.grid(True)                 # 그리드
plt.show()

```

### 3.1.7. 그래프 여러 개 보여주기

```python
plt.figure(figsize = (10, 3))                    # figure 설정
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)  # 그래프 간격

for i0 in range(6):
    plt.subplot(2, 3, i0 + 1)                     # 그래프 묘사의 위치 지정
    plt.title(i0 + 1)
    plt.plot(x, fnQuadratic(x, i0), color = "k")
    plt.ylim(-20, 20)
    plt.grid(True)

plt.show()
```

# 3.2. 3차원 그래프 그리기

## 3.2.1. 이변수 함수

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def fnCubic(x0, x1):
    r = 2 * x0**2 + x1**2
    ans = r * np.exp(-r)
    
    return ans

# x0, x1에서 각각 f3 계산
xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = fnCubic(x0[i0], x1[i1])

print(np.round(y, 1))
```

## 3.2.2. 수치를 색으로 표현하기: pcolor

```python
plt.figure(3, figsize = (5, 3))
plt.jet() # plt.gray, plt.pink, plt.bone
plt.contour(y) # plt.pcolor(y), plt.surface(x0, x1, y)
plt.colorbar()
plt.show()
```

## 3.2.3. 함수의 표면을 표시: surface

```python
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1)
print(x0)
print("")
print(xx0)
print("")
print(xx1)
```

- 3차원 그래프를 만들기 위해 subplot 선언시 projection = "3d"으로 지정
- 그리고 그래프의 id를 나타내는 반환값을 ax로 저장
- figure 내에 subplot을 하나만 지정했지만, subplot(n1, n2, n, projection = "3d")로 여러 좌표계 지정가능
- surface를 포시하는 부분은 ax.plot_surface(xx0, xx1, y)
- 옵션인 rstride와 cstride에 자연수를 부여해 가로 및 세로로 몇 개의 선을 긋는 지 지정할 수 잇음
- 수가 적을수록 선의 간격이 조밀해짐
- alpha는 면의 투명도를 지정
- z축의 눈금은 기본상태로 표시하면 숫자가 겹치기 때문에 
- ax.set_zticks((0, 0.2)) 명령으로 z의 눈금을 0과 0.2으로 제한
- ax.view_int(인수1, 인수2)는 3차원 그래프의 방향을 조절
- 인수1은 상하 회전 각도, 0을 지정하면 옆에서 본 그래프, 90을 지정하면 위에서 본 그래프
- 인수2는 좌우 회전 각도, 양수는 시계방향, 음수는 반시계 방향으로 차트가 회전

```python
plt.figure(figsize = (5, 3.5))
ax = plt.subplot(1, 1, 1, projection = "3d")
ax.plot_surface(xx0, xx1, y, 
               rstride = 3,
               cstride = 3,
               alpha = 0.3,
               color = "blue", edgecolor = "black")
ax.set_zticks((0, 0.2))
ax.view_init(15, 30)
plt.show()
```
